**Relatório de Análise Preditiva de Churn**

**1. Metodologia**

Este relatório apresenta os resultados de uma análise preditiva de churn realizada utilizando um modelo de regressão logística. O objetivo do modelo é prever a probabilidade de um cliente cancelar seu contrato com base em uma série de variáveis, como Segmento,	Alunos_Cadastrados,	Alunos_Ativos,	Taxa_Alunos_Ativos,	Questoes_Feitas,	Professores_Cadastrados,	Professores_Ativos,	Taxa_Professores_Ativos.

Os dados utilizados para treinar e avaliar o modelo foram obtidos de uma planilha do Excel. Antes de serem utilizados, os dados foram pré-processados para verificar a existência de valores ausentes e duplicados, além de transformar variáveis categóricas em numéricas.

O modelo de regressão logística foi treinado utilizando uma divisão de 70/30 entre os dados de treino e teste. O modelo foi então avaliado em termos de acurácia, sensibilidade e especificidade.

Visando melhorar a acurácia obtida, foram criados mais dois modelos, um deles aplicando a normalização dos dados e o outro a sua padronização. Ambos não obtiveram resultados satisfatórios.

**2. Resultados**

O modelo de regressão logística obteve uma acurácia de 75% na base de teste. Isso significa que o modelo foi capaz de prever corretamente o status do cliente (ativo ou inativo) em 75% dos casos.

A sensibilidade do modelo foi de 80%, o que significa que o modelo foi capaz de identificar corretamente 80% dos clientes inativos. A especificidade do modelo foi de 70%, o que significa que o modelo foi capaz de identificar corretamente 70% dos clientes ativos.

**3. Conclusão**

Os resultados obtidos neste relatório indicam que o modelo de regressão logística é uma ferramenta útil para prever a probabilidade de um cliente cancelar seu contrato. O modelo pode ser utilizado para identificar clientes em risco de churn e tomar medidas para retê-los.

PRÉ-PROCESSAMENTO

```python
# Para a análise, o código do cliente é irrelevante. Vamos eliminá-lo do dataframe

df = df.drop(columns=['Cliente'])
```

```python
# A variável Segmento é categórica. Para usá-la no modelo, ela precisar ser decodificada em numérica

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Segmento'] = le.fit_transform(df['Segmento'])
```

CONSTRUÇÃO E AVALIAÇÃO DA MÁQUINA PREDIDIVA - MODELO I

```python
# Para este modedo, usei Regressão Logística

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report


# Determinando o target (variável alvo)
X = df.drop('Status_Cliente', axis=1)
y = df['Status_Cliente']

# Separação das bases de teste e treino (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

# Treinando o modelo
model = LogisticRegression()
model.fit(X_train, y_train)

# Fazendo as predições com a base de teste
y_pred = model.predict(X_test)

# AVALIAÇÃO - Apurando a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)
print('Acurácia: {:.2f}%'.format(accuracy*100))

# Gerando a matriz de comfusão
cm = confusion_matrix(y_test, y_pred)
print(cm)

report = classification_report(y_test, y_pred)

print(report)
```
